```python
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
lbl.fit(list(df["HomeTeam"].values))
df["HomeTeam_Code"]=lbl.transform(list(df["HomeTeam"].values))
df["AwayTeam_Code"]=lbl.transform(list(df["AwayTeam"].values))
lbl2=LabelEncoder()
lbl2.fit(list(df["FTR"].values))
df["FTR"]=lbl2.transform(list(df["FTR"].values))
```

from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='median')
imp.fit(df)
df=pd.DataFrame(data=imp.transform(df),columns=df.columns)

```python
x=df.drop(["FTR","HomeTeam","AwayTeam"],axis=1)
y=df["FTR"]
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model_1=LogisticRegression(penalty="l2")
model_1.fit(x_train,y_train)
```

```python
model_1.score(x_test,y_test)
```

```python
model_1.score(x_train,y_train)
```

```python
model_1.predict(x_test)
```

```python
y_test
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_2=GradientBoostingClassifier(n_estimators=100)
model_2.fit(x_train,y_train)
```

```python
model_2.score(x_test,y_test)
```

```python
model_2.score(x_train,y_train)
```

```python
from xgboost import XGBClassifier
model_3=XGBClassifier(seed=100,predictor='gpu_predictor')
model_3.fit(x_train,y_train)
```

```python
model_3.score(x_test,y_test)
```

```python
model_3.score(x_train,y_train)
```

```python
from sklearn.ensemble import StackingClassifier,AdaBoostClassifier,RandomForestClassifier,BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
stack_model=StackingClassifier(estimators=[('model_3',model_3),('model_2',model_2),('model_1',model_1)],final_estimator=model_1)
```

```python
x2=df.drop(["FTR","FTHG","FTAG","HS","AS","HR","AR","HY","AY","HC","AC","HF","AF","HST","AST","HomeTeam","AwayTeam"],axis=1)
y2=df["FTR"]
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=0.2,random_state=42)
stack_model.fit(x2_train,y2_train)
```
